```python
import pandas as pd

# Carregar os arquivos CSV
file1 = r'D:\Machine-Learning-Empirical-Evaluation\TrainedModels\Unbalanced_consolidated_metrics.csv'
file2 = r'D:\Machine-Learning-Empirical-Evaluation\TrainedModels\Balanced_consolidated_metrics.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
# Verificar os tipos de dados de cada coluna
df1_dtypes = df1.dtypes
df2_dtypes = df2.dtypes

df1_dtypes, df2_dtypes
```

```python
# Selecionar apenas as colunas numéricas
numeric_columns = df1.select_dtypes(include=['float64']).columns

# Calcular a diferença entre as métricas nas colunas numéricas
difference_df = df2[numeric_columns] - df1[numeric_columns]

# Adicionar a coluna 'Model' ao DataFrame de diferença
difference_df['Model'] = df1['Model']

# Salvar o resultado em um novo arquivo CSV
output_file = r'D:\Machine-Learning-Empirical-Evaluation\Difference_consolidated_metrics.csv'
difference_df.to_csv(output_file, index=False)

output_file
```

```python
df3 = pd.read_csv(output_file)
df3.head()
df3 = df3.drop('Accuracy_std', axis=1)
df3 = df3.drop('Precision_std', axis=1)
df3 = df3.drop('Recall_std', axis=1)
df3 = df3.drop('F1_score_std', axis=1)
df3 = df3.drop('Roc_auc_std', axis=1)
df3
```
